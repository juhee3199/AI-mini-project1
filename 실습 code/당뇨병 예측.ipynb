```
import numpy as np
import torch
import torch.optim as optim
import pandas as pd

xy=pd.read_csv('2020.AI.diabetes-train.csv', header=None)
#xy=np.loadtxt('2020.AI.diabetes-train.csv', delimiter=",", dtype=np.float32, skiprows=1, usecols=range(0,9))


x_data=xy.loc[1:,0:7]
y_data=xy.loc[1:,8].squeeze()

x_data=np.array(x_data)
y_data=np.array(y_data)

x_train=torch.FloatTensor(x_data)
y_train=torch.LongTensor(y_data)

print(x_train[:5])
print(y_train[:5])


print(x_train.shape)
print(y_train.shape)
```

```
import torch.nn.functional as F

nb_class=2
nb_data = len(y_train)

W=torch.zeros((8,nb_class), requires_grad=True)
b=torch.zeros(1, requires_grad=True)

optimizer = optim.SGD([W,b], lr=0.0001)

nb_epochs=10000
for epoch in range(nb_epochs+1):
  
  hypothesis = F.softmax(x_train.matmul(W)+b, dim=1)

  y_one_hot=torch.zeros(nb_data, nb_class)
  y_one_hot.scatter_(1,y_train.unsqueeze(1),1)
  cost=(y_one_hot * -torch.log(F.softmax(hypothesis, dim=1))).sum(dim=1).mean()

  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  if epoch % 10==0:
    print('Epoch {:4d}/{} cost: {:.6f} '.format(
        epoch, nb_epochs, cost.item()
    ))
```
