## [1] 판다스를 이용한 동물예측 데이터 로더

```
import numpy as np
import torch
import torch.optim as optim
import pandas as pd

torch.manual_seed(1)

! curl 'https://raw.githubusercontent.com/sejongresearch/2020.Spring.AI/master/HW/zoo_class.csv' -o 'zoo_class.csv'
print('Done..\n')

xy=np.loadtxt('zoo_class.csv',delimiter=',', dtype=np.float32)
    #numpy를 이용해 읽어오는 함수.
    #numpy를 이용해서 읽어오게되면 헤더까지 같이 읽어오게된다.
    # 그래서 pandas를 이용하기.

pandas_xy=pd.read_csv("zoo_class.csv", header=None); 
                               #헤더 none은 데이터에 따라 다르다.
```

```
xy        #numpy로 읽어온 데이터. 분석하기 어려운 데이터
```

```
#pandas로 읽어온 데이터. 
# 가시성이 높음.데이터의 크기를 알 수 있다. row x columns

pandas_xy 
```

```
# 데이터 헤더정보 확인하기   헤더=columns
    # 보통 csv로 만들어서 배포시, 헤더에 columns 정보가 있다. 하지만 이 데이터에는 없음.

pandas_xy.keys()
```

```
pandas_xy.index  # 101 rows
```

```
# 클래스 개수 확인하기

pandas_xy[16].max()
#pandas_xy.iloc[:,16].max()

#pandas_xy[16].min() 
#min=0, max=6. 0~6까지 7개의 클래스
```

```
#인덱스에 따른 row의 정보를 받아오기.

pandas_xy.iloc[0]  
```

```
# 헤더의 값을 가지고 cols의 정보가져오기

pandas_xy[0]
```

```
# 구역의 값 가져오기
# loc[인덱스의 범위, 헤더의 범위]

pandas_xy.loc[0:10,0:5]
```

```
# 예제 파싱하기

x_data=pandas_xy.loc[:,:15]
y_data=pandas_xy.loc[:,16]

x_data    #cols 0~15
```

```
y_data  # 16cols
```

```
# 이전에 배운것을 활용하기 위해서는 numpy로 변경해야 함.

x_data=np.array(x_data)
y_data=np.array(y_data)
x_data
```
