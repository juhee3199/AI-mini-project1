### 데이터 입출력 정의



```
# CPU 모드
import torch
```

```
torch.manual_seed(777)  #랜덤 씨드를 고정시켜서 매번 값이 같게 나오도록한다.
```

```
X = torch.FloatTensor([[0,0], [0,1], [1,0],[1,1]])
Y = torch.FloatTensor([[0],[1],[1],[0]])
```

```
X
```

```
Y
```

### NN 모델 정의

```
# 모델 정의
# Layer 정의

# torch.nn.Linear : 한 개의 logistic regression 
# (2,1,bias=True) : 2개의 입력 -> 1개의 출력을 의미.

linear = torch.nn.Linear(2,1,bias=True) # Fully Connected Layer
sigmoid = torch.nn.Sigmoid() 
```

```
model = torch.nn.Sequential(linear,sigmoid)  #linear와 sigmoid를 연결
```

```
model
```

### 모델 학습

```
# 모델 학습 환경 설정: loss function, optimizer설정 등등

loss = torch.nn.BCELoss()   # Binary Cross Entropy Loss
optimizer = torch.optim.SGD(model.parameters(),lr=1)    #모델의 파리미터를 학습, lr=1
```

```
for stop in range(10000):
    
    # 그래디언트 초기화
    optimizer.zero_grad()
    # Forward 계산
    hypothesis = model(X)       # 예측값 y를 구함.

    # Error 계산
    cost = loss(hypothesis, Y)  # 예측값과 실제 y값의 차이.에러
    # Backward 계산 
    cost.backward()             # backward를 계산
    # 가중치 갱신
    optimizer.step()            # 에러를 기준으로 w,b를 갱신

    if stop % 100 == 0:
        print(stop, cost.item())
```

### 모델 평가

#### --> 퍼셉트론에서는 XOR문제를 해결할 수 없으므로 어떤 방법을 사용해도 정확도는 1이 나오지 않는다.

```
## w,b 평가

# required_grad = true는 학습중임을 의미.
# required_grad = false 모델을 평가중임을 의미. 학습하지 않음. 

with torch.no_grad(): # 임시로 required_grad = false로 설정하는 것과 같다.

    hypothesis = model(X)
    predicted = (hypothesis > 0.5).float()    # 1 or 0으로 바꾸기
    accuracy = (predicted == Y).float().mean()  
    print('\n Hypothesis: ', hypothesis.numpy(), '\n Correct: ', predicted.numpy(), '\n Accuracy: ', accuracy.item())

# 퍼셉트론에서는 XOR문제를 해결할 수 없으므로 어떤 방법을 사용해도 정확도는 1이 나오지 않는다.
```
